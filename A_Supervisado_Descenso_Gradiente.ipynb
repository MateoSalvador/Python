```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
def func(th):
    x = th[0]
    y = th[1]
    val =np.sin(1/2 * x**2 -1/4 * y**2 + 3) * np.cos(2*x + 1 - np.e**y)
    return val
```

```python
func([5,3])
```

```python
#Crear variables x,y,z
x = np.linspace(-2,2,100) #100 valores
y = np.linspace(-2,2,100) # 100 valores
z = np.zeros([len(x), len(y)])
```

```python
for ix, xval in enumerate(x):
    for iy,yval in enumerate(y):
        z[iy,ix] = func([xval,yval])
```

```python
plt.contourf(x,y,z,100)
plt.colorbar()
plt.show()
```

```python
# Plot the surface
from matplotlib import cm
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x,y,z,cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
##Clcula de la derivada 

 

theta = np.random.rand(2) * 4 - 2
theta
plt.contourf(x,y,z,100)
plt.colorbar()
plt.plot(theta[0],theta[1],marker='.',markersize='20',c='c')

 

#Cculo de las derivadas parciales

 

theta1 = np.copy(theta)

 

h = 0.001 #pequeo valor (incrmento)

 

gradi = np.zeros(len(theta))

 

alpha = 0.01

 

for _ in range(1000): #iterativa

 

    #Calculo de pequeas diferencias de los parmetros de la funcin (derivada), incremento
    for ith, th_val in enumerate(theta): #recorrer los componentes del vector
    
        theta1 = np.copy(theta) #asignar el valor del vector theta
        
        theta1[ith] = theta1[ith] + h 
    
        derivada = (func(theta1) - func(theta))/h  #derivada parcial
    
        gradi[ith] = derivada

 

    theta = theta - alpha * gradi
    #print(func(theta))
    plt.plot(theta[0],theta[1],marker='.',markersize='10',c='r')
    
plt.plot(theta[0],theta[1],marker='.',markersize='10',c='r')
plt.show()
```
