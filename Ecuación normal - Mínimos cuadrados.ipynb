```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
#Varialbe de entrada x
x = boston.data[:,5]
print(x) 
```

```python
#Variable de salida y
y = boston.target
print(y)
```

```python
#Análisis del dataset
#x_val = np.array(x)
y_val = np.array(y)
plt.scatter(x,y,alpha=0.7)
plt.xlabel('Num Dormitorios')
plt.ylabel('Precio')
plt.show()
```

Minimización
$\theta = (X^{T}X)^{-1}X^{T}Y$

```python
#Añadir una columna de 1s que representa el término indpendiente
x_val = np.array([np.ones(int(x_val.size)),x_val]).T
```

```python
x_val
```

```python
# Multiplicación vectorial (matrcial)
theta = np.linalg.inv((x_val.T @ x_val)) @ x_val.T @ y_val 
```

```python
theta
```

```python
plt.plot([4,9],[theta[0]+theta[1]*4, theta[0]+theta[1]*9])
```

```python
plt.scatter(x,y,alpha=0.5)
plt.plot([4,10],[theta[0]+theta[1]*4, theta[0]+theta[1]*9],c='r')
plt.xlabel('Num Dormitorios')
plt.ylabel('Precio')
plt.show()
```

```python
plt.scatter(x,y,alpha=0.5)
plt.plot([4,10],[theta[0]+theta[1]*4, theta[0]+theta[1]*10],c='r')
plt.xlabel('Num Dormitorios')
plt.ylabel('Precio')
plt.show()
```

```python
plt.scatter(x,y,alpha=0.5)
plt.plot([4,15],[theta[0]+theta[1]*4, theta[0]+theta[1]*15],c='r')
plt.xlabel('Num Dormitorios')
plt.ylabel('Precio')
plt.show()
```
