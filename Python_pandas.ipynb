### Obtención del tipo de datos de un DataFrame
El tipo de dato del valor almacenado en una celda de un DataFrame determina el **tipo de operaciones** que se puede realizar sobre él. En python se usa el método *type*, mientras que en Pandas se usa la propiedad *dtype* sobre una columna o fila y *dtypes* sobre el DataFrame completo.  

| Tipo de dato en Pandas  | Tipo de dato en Python | Descripción  |
| ------------- | ------------- | --------|
| object  | string  | Cadena de texto. Si contiene tipos mixtos (números y cadenas), será asignado un object/string.|
| int64  | int  | Números (cada uno será almacenado con 64 bits)|
| float64 | float| Números flotantes (cada uno será almacenado con 64 bits)|
| datetime64/timedelta | datetime| Valores con datos de tiempo|

```python
type(traffic_df)
traffic_df['json.apk'].dtype
traffic_df.dtypes
```

### Cambio de valores a ciertas filas (basado en condiciones)
La propiedad loc[] de un DataFrame permite **acceder** a un grupo de filas o columnas mediante sus etiquetas o un array de Booleanos. Una vez identificadas las celdas de interés, se puede asignar cierto valor. 

```python
traffic_df.loc[traffic_df['json.port'] == '443', 'protocolo'] = 'HTTPS'
```

En el siguiente ejemplo creamos un DataFrame con los registros cuyo campo 'json.PII' contiene alguno de los siguientes valores: ["Contact_Address_Book", "Contact_E_Mail_Address",  "Contact_Phone_Number",  "Contact_Name"]. Para ello, crearemos una columna auxiliar 'pd' a la que asignaremos el valor 'yes', únicamente cuando el campo 'json.PII' coincida con los ya indicados. Así:

```python
pd_lst = ["Contact_Address_Book", "Contact_E_Mail_Address", "Contact_Phone_Number", "Contact_Name"]
traffic_df['pd'] = 'No'
for pii in pd_lst:
    traffic_df.loc[(traffic_df['json.PII']==pii), 'pd'] = 'Yes'

#traffic_df    
traffic_df[traffic_df['pd'] == 'Yes'].head(10)
```

### Eliminación de duplicados
El método drop_duplicates() permite eliminar las filas duplicadas. Uno de los argumentos de entrada es un vector con las etiquetas de las columnas en base a las cuales se determina si la fila es duplicada. En el siguiente ejemplo usamos este método para encontrar el número total de aplicaciones móviles evaluadas:

```python
total_apps = traffic_df.drop_duplicates(['json.apk']).shape[0]
print('Número de apps evaluadas: ', total_apps)
```

### Celdas vacías en Pandas
En Pandas, a las celdas vacías se asigna el valor 'NaN' (Not a Number). NaN signica que no se ha definido un valor. Considere el siguiente ejemplo:

```python
import numpy as np
aux_df = pd.DataFrame({'country': ['Ecuador', 'Perú', 'Colombia', np.nan], 'population':[10, np.nan, 10, 10]})
aux_df
```

Los valores NaN tienen utilidad para algunas operaciones, por ejemplo para promediar una columna Python ignorará las celdas vacías (no las considera en el cálculo). 

```python
aux_df.mean()
```

Mostrar las filas que tienen valores vacíos (NaN) en alguna columna en particular

```python
aux_df[aux_df.population.isna() | aux_df.country.isna()]
```

```python
aux_df['population'] = aux_df['population'].fillna(0)
aux_df.mean()
```

Eliminar las filas que tienen valores vacíos (NaN) en alguna columna en particular

```python
missing_removed_aux = aux_df.dropna(subset=['country'])
missing_removed_aux
```

### Agrupamiento de datos 
Podemos agrupar (y contar) filas en base a valores similares de las columnas. En el ejemplo, queremos obtener los dominios predominantes a donde se filtran los datos personales. Para ello, ordenamos (groupby), contamos (count) y ordenamos (sort_values) los datos del DataFrame. 

```python
top_domains = traffic_df.drop_duplicates(['json.apk', 'SLD']).groupby('SLD').count().sort_values(['json.apk'], ascending=False)[['json.apk']]
top_domains
```

```python
top_domains = top_domains.reset_index()
top_domains
```

### Visualización de los datos
Para la visualización de datos se puede usar la biblioteca Seaborn. Seaborn está desarrollados obre matplotlib y provee una interfaz de alto de nivel para crear diferentes gráficos estadísiticos. Seaborn nos ayuda a explorar y entender nuestros datos y operan sobre los Dataframes que contienen los datos; internamente ejecutan las operaciones necesarias para visualizar los datos de acuerdo a los datos escogidos. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="white")
fig, ll = plt.subplots(1, 1, figsize = (13,4))
ax = sns.barplot(x="SLD", y="json.apk", data=top_domains.head(20), dodge=False)
ax.set(xlabel="Domain", ylabel="Number of apps")
plt.xticks(rotation=90)
plt.savefig('top_domains.svg', bbox_inches='tight', pad_inches=1)
plt.show()
```
